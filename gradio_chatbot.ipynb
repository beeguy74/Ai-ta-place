```python
import gradio as gr
import random
import time
from typing import List, Tuple

messages = [
    {
        "role": "system",
        "content": "You are a chatbot who is going to help user administrative task. You'll make the call pretending to be a user. Ask the user for the required data.",
    },
]

with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    msg = gr.Textbox()
    clear = gr.ClearButton([msg, chatbot])

    def convert_to_chat_history(messages):
        chat_history = []
        for i in range(1, len(messages), 2):
            chat_history.append([messages[i]["content"], messages[i+1]["content"]])
        return chat_history

    def respond(message, chat_history):
        global messages
        message_d = {"role": "user", "content": message}
        messages.append(message_d)
        messages = pipe(messages, max_new_tokens=128)[0]['generated_text']  # Print the assistant's response
        # chat_history.append((message, bot_message))
        return "", convert_to_chat_history(messages)

    msg.submit(respond, [msg, chatbot], [msg, chatbot])

demo.launch(debug=True)
```
